# Process and Visualize Categorical Data with Pandas and Seaborn


* Utilize the `.info()` and `.select_dtypes()` methods in Pandas to identify and explore categorical data within a dataset.
*  enerate and customize count plots for each categorical column using Seaborn to visually analyze data distribution and frequency.
* Interpret visual data representations, gaining insights into different categorical variables to support more informed data-driven decisions

```python
data.info()
```

```python
catcols = data.select_dtypes(include = ["object", "category"]).columns
```

```python
for c in catcols:
  plt.figure (figsize= (6,4))
  sns.countplot(data = data, x = c)
  plt.title(f"countplot for {c}")
  plt.xticks(rotation = 45)
```

```python
numfeat = data.select_dtypes(include = ['float64', "int"]).columns.difference(['species'])


for feat in numfeat:
  plt.figure(figsize = (6,4))
  sns.boxplot(data = data, x = "species" , y = feat)
  plt.title(f"distribution for {feat} by species")
  plt.xlabel ("species")
  plt.ylabel(feat)
  plt.xticks(rotation = 45)
  plt.show()
```

```python
sns.kdeplot(data=data, x="flipper_length_mm", hue="species", multiple="stack");
```
